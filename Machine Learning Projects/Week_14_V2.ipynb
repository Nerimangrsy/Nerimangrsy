Task-14A: Please provide a line plot with the use of sin(y) and cos(y) this time and plot it properly on the x-y axis. Please use legends to illustrate what they are and please do not use the default setting for the linetypes and colors, but instead, specify them on your own.


```
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```
y = np.linspace(-np.pi, np.pi, 256, endpoint=True)

y1 = np.cos(y)
y2 = np.sin(y)

plt.plot(y1, y, color='blue', linestyle='dotted', label='cos(y)')
plt.plot(y2, y, color='red',  linestyle='dashed', label='sin(y)')
plt.title('A simple line plot')
plt.xlabel('sin(y) and cos(y)')
plt.ylabel('y')
plt.legend();
```

Task-14B: Please generate a random dataset of 27 people in a class which is composed of their height, width, and weight. You need to provide a scatter plot of these people by using these three parameters in the same order (x, y, size of the circle). Please do not use the default setting for the linetypes and colors, but instead, specify them on your own.

```
rng = np.random.RandomState(0)
h = rng.randn(27)  #height
wd = rng.randn(27)   #width
wg = 1000 * rng.rand(27)  #weight
colors = rng.rand(27)

plt.scatter(h, wd, alpha=0.3, 
            s= wg, c=colors,
            cmap='viridis')
plt.colorbar();  # show color scale
```

Task-14C: Please provide a scatter plot for the 7 big cities of Turkey similar to the cities of California. Please include the latitude and longitude, population and area of these cities (rough estimations are accepted). I would like to see a similar plot just like the one given below:

```
import pandas as pd
from matplotlib.pyplot import figure
data={'City': ['İstanbul', 'Eskişehir', 'İzmir', 'Bursa', 'Ankara', 'Trabzon', 'Adana'],
      'Population': [ 15519267, 887475, 4367251, 3056120, 5639076, 808974 , 2237940],
      'Area':[5461, 13960, 11891, 10813, 25632, 13844, 4628],
      'longitude': [28.97696, 30.520556, 27.12872, 29.063448 , 32.85411, 39.7191, 35.3324 ],
      'latitude': [41.00527, 39.776667, 38.41885, 40.266864, 39.92077, 41.0061, 36.9719 ]}
df=pd.DataFrame(data)
df
```

```
lat, lon = df['latitude'], df['longitude']
population, area = df['Population'], df['Area']

# Scatter the points, using size and color but no label
plt.scatter(lon, lat, label=None,
            c=np.log10(population), cmap='jet',
            s=area/10, linewidth=0, alpha=0.7)
plt.axis(aspect='equal')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3,7)
plt.show()
```
